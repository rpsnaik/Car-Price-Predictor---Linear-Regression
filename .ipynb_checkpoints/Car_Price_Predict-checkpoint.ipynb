```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
carprice_data = pd.read_csv("./CarPrice_Assignment.csv")
```

```python
carprice_data.head()
```

```python
sns.heatmap(carprice_data.corr())
```

```python
carprice_data['model'] = [x.split()[0] for x in carprice_data['CarName']]
carprice_data['model'] = carprice_data['model'].replace({'maxda': 'Mazda', 'nissan': 'Nissan', 'porcshce': 'Porsche', 'toyouta': 'Toyota', 
                            'vokswagen': 'Volkswagen', 'vw': 'Volkswagen'})

carprice_data= carprice_data.drop(['car_ID','CarName'], axis=1)
```

```python
carprice_data.head()
```

```python
carprice_data_copy = carprice_data
carprice_data = carprice_data.drop(["price"], axis=1)
carprice_data.head()
```

```python
X = carprice_data.iloc[:].values
from sklearn.preprocessing import LabelEncoder

labelEncoder = LabelEncoder()
X[:,1] = labelEncoder.fit_transform(X[:,1])
X[:,2] = labelEncoder.fit_transform(X[:,2])
X[:,3] = labelEncoder.fit_transform(X[:,3])
X[:,4] = labelEncoder.fit_transform(X[:,4])
X[:,5] = labelEncoder.fit_transform(X[:,5])
X[:,6] = labelEncoder.fit_transform(X[:,6])
X[:,12] = labelEncoder.fit_transform(X[:,12])
X[:,13] = labelEncoder.fit_transform(X[:,13])
X[:,15] = labelEncoder.fit_transform(X[:,15])
X[:,23] = labelEncoder.fit_transform(X[:,23])


# carprice_data = pd.get_dummies(carprice_data, drop_first=True)
# carprice_data.head()
```

```python
X
```

```python
Y = carprice_data_copy['price']
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 80)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
